### 문제 1
‘students.txt’ 파일을 읽어서 딕셔너리로 학생들의 정보를 저장하라.  
해당 파일은 이름, 국어점수, 영어점수, 수학점수 순으로 되어있으며, 각 컬럼의 구분자는 tab(\t)으로 이루어져있다.  
이름을 key로 하고, 국어점수, 영어점수, 수학점수를 순서대로 담은 리스트를 value로 하는 딕셔너리로 저장하라.

+ read 이용

```python
students = {}
```

```python
file_name = 'students.txt'
file_path = file_folder + file_name
```

```python
# 정답을 작성해 주세요.
with open(file_path, 'r') as f:
    data = f.read()
    # print(data)
    # print('-----')
    lines = data.split('\n')
    # print(lines)

    # vue = lines[0].split('\t')
    # print(vue)
    # jk = lines[1].split('\t')
    # print(jk)
    # jm = lines[2].split('\t')
    # print(jm)

    for student in lines:
        # print(student)
        s = student.split('\t')
        # print(s)
        # students['뷔'] = ['50', '70', '91]
        # print(s[0])
        # print([s[1], s[2], s[3]])
        # print(s[1:])
        students[s[0]] = s[1:]
```

### 문제 2
저장한 딕셔너리를 토대로 학생들의 평균을 구하라.  
전체 학생의 국어, 영어, 수학 점수의 평균을 구해서 각각 kor, eng, math라는 변수에 저장하고, 출력하라.

```python
students
```

```python
kor = 0
eng = 0
math = 0
```

```python
# 정답을 작성해 주세요.

# (stundet['RM'][0] + stundet['뷔'][0] +.... ) / 7
# print(students.values())

for v in students.values():
    print(v[0])
    # kor = kor + v[0]
    kor += int(v[0])
    eng += int(v[1])
    math += int(v[2])

print(kor)
# result_kor = kor / len(students)
kor = kor / len(students)
eng /= len(students)
math /= len(students)

# print(kor)


```

```python
print('학생들의 평균은 아래와 같습니다.')
print('국어 : ' + str(kor))
print('영어 : ' + str(eng))
print('수학 : ' + str(math))
```
