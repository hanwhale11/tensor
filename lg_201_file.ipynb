### 1-2. with문을 활용하여 파일 쓰기

```python
file_path_2 = ex_dir + 'new2.txt'
# file_path_2 = '/content/drive/My Drive/20210607_11_lg/new2.txt'
with open(file_path_2, 'w') as f :
    data = '이렇게도 작성이 가능합니다. close를 따로 안해도 돼요.'
    f.write(data)
```

### 2. 파일 읽기
- read() : 파일 전체를 한 번에 읽어서 string으로 반환
- readline() : 파일을 한 줄만 읽어서 string 으로 반환 (줄바꿈문자 \n)
- readlines() : 파일을 줄 단위로 모두 읽어서 list 로 반환

```python
with open(file_path_1, 'r') as f:
    lines = f.read()
    print(lines)
    print(type(lines))
```

```python
with open(file_path_1, 'r') as f:
    line = f.readline()
    print(line)

    line2 = f.readline()
    print(line2)

    print(type(line))
```

```python
with open(file_path_1, 'r') as f:
    while True :
        line = f.readline()
        if not line : break
        print(line)
```

```python
with open(file_path_1, 'r') as f:
    lines = f.readlines()
    print(lines)
    print(type(lines))
```
