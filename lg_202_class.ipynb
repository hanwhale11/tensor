## 인스턴스 변수와 인스턴스 메소드 예시

```python
class Account():
    def make_account(self) :
        self.balance = 0

    def deposit(self, money) :
        self.balance += money

    def draw(self, money) :
        self.balance -= money
```

```python
a1 = Account()
a1.make_account()
print(a1.balance)
a1.deposit(1000)

print(a1.balance)
```

```python
a2 = Account()
a2.make_account()
a2.deposit(5000)
print(a2.balance)
```

```python
Account.balance
```
