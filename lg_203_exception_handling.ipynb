### try ~except

```
try: 
    예외를 유발할 수 있는 구문 
except <예외 종류>: 
    예외 처리를 수행하는 구문
```

````python
try: 
    print(123 / 0)
except ZeroDivisionError:
    print('0으로 나눌 수 없습니다.')
````

````python
lst = [1, 2, 3]
try: 
    print(lst[3])
except IndexError:
    print('옳지 않은 인덱스 값입니다')
    print(f'리스트의 길이는 {len(lst)}입니다.')
````

````python
try: 
    print(123 / 0)
except ZeroDivisionError as e:
    print('에러정보: ', e)
    print('0으로 나눌 수 없습니다.')
````

````python
try: 
    print(open('./wrong_path/error.txt', 'r'))
except ZeroDivisionError:
    print('해당 경로에 파일이 존재하지 않습니다.')
````

````python
try: 
    print(open('./wrong_path/error.txt', 'r'))
except FileNotFoundError:
    print('해당 경로에 파일이 존재하지 않습니다.')
````
