#### Pop Quiz

여기는 동네 유명한 빵집이다. 사람들에게 먼저 온 순서대로 번호표를 나누어주려고 한다. 번호표를 나누어주는 함수를 작성해보자.  
함수는 사람이름으로 되어 있는 리스트를 받아서 "대기번호 x번 : 사람이름" 를 화면에 출력하고 (번호표, 사람이름)을 원소로하는 이루어진 리스트를 반환한다. 
- input : 리스트
- output : 리스트

```python
people = ['지민', '뷔', 'RM', '제이홉']
```

```python
def func1(line) :
    new_lines = []
    i = 1
    for x in line :
        print(f"대기번호 {i}번 : {x}")
        new_lines.append((i, x))
        i += 1
    return new_lines
```

```python
lines = func1(people)
```

```python
def func1_with_enu(line) :
    new_lines = []
    for idx, val in enumerate(line):
        print("대기번호" + str(idx+1) + ":" + val)
        new_lines.append((idx+1,val))
    return new_lines
```

# map
- map(함수, 자료형)
-	리스트, 튜플, 스트링 등 자료형 각각의 원소에 동일한 함수를 적용

```python
items = [1, 2, 3, 4, 5]
```

```python
squared = []
for i in items :
    # print(i)
    result = i * i 
    # print(result)
    squared.append(result)
```

```python
print(squared)
```

```python
squared_map = list(map(lambda x : x**2, items))
```

```python
print(squared_map)
```

lambda와 map을 이용하여 items의 요소들을 string(문자)로 바꾸는 것을 짜봅시다.

```python
str_items = list(map(lambda x : str(x), items))
```

```python
list_3 = []
for x in range(10, 21) :
    # print(x)
    if x % 2 == 0 :
        # print(x)
        list_3.append(x)
print(list_3)
```

## 3) for 문 +  if / else 문
1부터 10 사이의 숫자들 중 홀수이면 제곱수를, 짝수이면 세제곱수를 저장하라.

```python
list_4 = []
for x in range(1, 11) :
    if x % 2 == 1 :
        list_4.append(x ** 2)
    else :
        list_4.append(x ** 3)
```

```python
list_4
```

```python
lc_4 = [x ** 2 if x % 2 == 1 else x ** 3 for x in range(1 , 11)]
```

```python
lc_4
```
